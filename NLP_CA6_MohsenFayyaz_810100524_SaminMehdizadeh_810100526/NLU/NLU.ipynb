```python
import json
with open('data.intents') as f:
    intents = f.read()
intents = json.loads(intents)

with open('data.slots') as f:
    slots = f.read()
slots = json.loads(slots)
```

```python
print('Count of Intents:',len(intents.keys()))
print('Count of Slots:',len(slots.keys()))
```

```python
print('Intent Types:')
print(list(intents.values()))
print('Slot Types:')
print(list(slots.values()))
```

```python
with open('1.0/data/fa-IR.jsonl','r') as f:
    data_list = list(f)

intents_count = {'test' : {intent: 0 for intent in intents.values()},
                 'train' : {intent: 0 for intent in intents.values()},
                 'dev' : {intent: 0 for intent in intents.values()}};
slots_count = {'test' : {slot: 0 for slot in slots.values()},
                 'train' : {slot: 0 for slot in slots.values()},
                 'dev' : {slot: 0 for slot in slots.values()}};
for json_str in data_list:
    result = json.loads(json_str)
    partition = result['partition']
    intent = result['intent']
    slot_method = result['slot_method']
    for method in slot_method:
      slots_count[partition][method['slot']] = slots_count[partition][method['slot']]+1
    intents_count[partition][intent] = intents_count[partition][intent] +1
```

```python
print("intents count")
print(json.dumps(intents_count, indent=2))
```

```python
import matplotlib.pyplot as plt
plt.plot(list(intents_count['test'].values()),label='test')
plt.plot(list(intents_count['train'].values()),label='train')
plt.plot(list(intents_count['dev'].values()),label='dev')
plt.title('intents')
plt.legend()
plt.show()
```

```python
plt.plot(list(slots_count['test'].values()),label='test')
plt.plot(list(slots_count['train'].values()),label='train')
plt.plot(list(slots_count['dev'].values()),label='dev')
plt.title('slots')
plt.legend()
plt.show()
```

```python
print("slots count")
print(json.dumps(slots_count, indent=2))
```
